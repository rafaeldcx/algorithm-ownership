```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Lista de arquivos CSV
csv_files = ["Amazon-FreeRTOS.csv",
             "Contiki.csv",
             "Contiki-ng.csv",
             "FreeRTOS.csv",
             "Mbed-OS.csv",
             "Apache-NuttX.csv",
             "Riot.csv",
             "RT-Thread.csv",
             "Zephyr.csv"]

# Definir o tamanho do grfico
plt.figure(figsize=(30, 20))  # Aumentar o tamanho da figura

# Cores para cada linha
colors = sns.color_palette("hsv", len(csv_files))

# Loop pelos arquivos CSV
for i, file in enumerate(csv_files):
    data = pd.read_csv(file, encoding='utf-8')
    data['Descrio'] = pd.to_numeric(data['Descrio'], errors='coerce')
    
    # Plotar a ECDF para cada arquivo
    sns.ecdfplot(data=data, x='Descrio', label=file.split('.')[0], color=colors[i], linewidth=3.0)

# Ajustes do grfico
plt.ylim(0, 1)  # Ajustar a escala do eixo y
plt.yticks(np.arange(0, 1.1, 0.1))  # Ajustar os ticks do eixo y para que sejam de 0.1 em 0.1

plt.xlabel('Nmero de Mdulos por Responsvel', fontsize=25)  # Aumentar o tamanho do rtulo do eixo x
plt.ylabel('Proporo', fontsize=25)  # Adicionar e aumentar o tamanho do rtulo do eixo y

# Ajustar a escala do eixo x
plt.xlim(0, 100)  # Ajuste os valores conforme necessrio

# Ajustar os ticks do eixo x para que sejam de 5 em 5
plt.xticks(np.arange(0, 101, 5), fontsize=20)  # Aumentar o tamanho dos ticks do eixo x
plt.yticks(fontsize=20)  # Aumentar o tamanho dos ticks do eixo y

# Adicionar a legenda no lado direito
plt.legend(title='Projetos', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20, title_fontsize=25)

# Exibir o grfico
plt.show()
```
