```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python

# Lista de arquivos CSV contendo a contagem de responsáveis por arquivo para diferentes projetos
csv_files = [
    "Amazon-FreeRTOS.csv",
    "Contiki.csv",
    "Contiki-ng.csv",
    "FreeRTOS.csv",
    "Mbed-OS.csv",
    "Apache-NuttX.csv",
    "Riot.csv",
    "RT-Thread.csv",
    "Zephyr.csv"
]

# Definir o tamanho do gráfico
plt.figure(figsize=(30, 20))

# Cores para cada linha
colors = sns.color_palette("hsv", len(csv_files))

# Loop pelos arquivos CSV
for i, file in enumerate(csv_files):
    data_responsaveis = pd.read_csv(file, encoding='utf-8')
    
    # Verificar se a coluna 'Descrição' está presente
    if 'Descrição' in data_responsaveis.columns:
        # Criar o gráfico ECDF para a contagem de responsáveis por arquivo
        sns.ecdfplot(data=data_responsaveis, x='Descrição', label=file.split('.')[0], color=colors[i], linewidth=3.0)
    else:
        print(f"A coluna 'Descrição' não foi encontrada no arquivo {file}. Verifique o nome da coluna.")

# Ajustes do gráfico
plt.ylim(0, 1)  # Ajustar a escala do eixo y
plt.yticks(np.arange(0, 1.1, 0.1))  # Ajustar os ticks do eixo y para que sejam de 0.1 em 0.1

plt.xlabel('Número de Responsáveis por Módulo', fontsize=25)  # Aumentar o tamanho do rótulo do eixo x
plt.ylabel('Proporção', fontsize=25)  # Adicionar e aumentar o tamanho do rótulo do eixo y

# Ajustar a escala do eixo x, se necessário
plt.xlim(0, 9)

# Ajustar os ticks do eixo x para intervalos regulares, ajustado para 1 em 1
plt.xticks(np.arange(0, 10, 1), fontsize=20)
plt.yticks(fontsize=20)  # Aumentar o tamanho dos ticks do eixo y

# Adicionar a legenda no lado direito
plt.legend(title='Projetos', loc='center left', bbox_to_anchor=(1, 0.5), fontsize=20, title_fontsize=25)

# Exibir o gráfico
plt.show()

```
